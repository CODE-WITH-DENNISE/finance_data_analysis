```python
import os
import pandas as pd

# Define the directories
directories = ['network 1', 'network2', 'network3']
output_directory = 'merged2'

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Function to check if a file is hidden
def is_hidden(filepath):
    return any(part.startswith('.') for part in filepath.split(os.sep))

# Get the list of all file names from the first directory
file_names = [f for f in os.listdir(directories[0]) if not is_hidden(f)]

# Iterate over each file name
for file_name in file_names:
    # List to hold dataframes
    dataframes = []
    
    # Read the same file from each directory and append it to the list
    for directory in directories:
        file_path = os.path.join(directory, file_name)
        if os.path.exists(file_path) and not is_hidden(file_path):
            try:
                df = pd.read_csv(file_path, encoding='latin1')  # Specify the encoding
                dataframes.append(df)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
    
    # Concatenate all dataframes
    if dataframes:
        merged_df = pd.concat(dataframes, ignore_index=True)
    
        # Save the merged dataframe to a new CSV file
        output_path = os.path.join(output_directory, file_name)
        try:
            merged_df.to_csv(output_path, index=False, encoding='utf-8')
        except Exception as e:
            print(f"Error writing {output_path}: {e}")

print("CSV files have been merged and saved.")


```

```python
import os
import pandas as pd

# Define the directory containing the merged CSV files
merged_directory = 'merged2'
output_file = 'afinal_merged.csv'

# Get the list of all CSV file names in the merged directory
file_names = [f for f in os.listdir(merged_directory) if f.endswith('.csv')]

# List to hold dataframes
dataframes = []

# Read each CSV file and append it to the list
for file_name in file_names:
    file_path = os.path.join(merged_directory, file_name)
    try:
        df = pd.read_csv(file_path, encoding='latin1')  # Specify the encoding
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all dataframes
if dataframes:
    final_merged_df = pd.concat(dataframes, ignore_index=True)
    
    # Save the final merged dataframe to a new CSV file
    try:
        final_merged_df.to_csv(output_file, index=False, encoding='utf-8')
        print(f"All CSV files have been merged and saved to {output_file}.")
    except Exception as e:
        print(f"Error writing {output_file}: {e}")
else:
    print("No dataframes to merge.")
```
